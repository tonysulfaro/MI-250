## Suggested Schedule
Every module can be worked through at your own pace within the timeframe of the module. Modules are 9 days long from Monday to Tuesday of the following week. Monday is always the start day of a new module so there is a two day overlap between modules. The only component that must be completed by a particular day is the discussion as the other students will be required to respond to several submissions. Posting within by the intended date will allow others adequate time to respond to your post. Discussions will be marked by an asterix (*) below under the date by which a post is required.

**Monday - Wednesday (Day 1 - Day 3)** 
- Labs
- Readings
- Practices

**Wednesday (Day 3)**
- Post Discussion

**Thursday (Day 4)**
- Respond to Discussions

**Friday - Tuesday (Day 5 - Day 9)**
- Exercises
